```python
import pandas as pd

df = pd.read_csv("../data/raw/products_raw.csv")
```

```python
df = df[
    [
        "parent_asin",
        "title",
        "description",
        "main_category",
        "price",
        "average_rating",
    ]
]
```

```python
df = df.drop_duplicates(subset="parent_asin")
```

```python
df["description"] = df["description"].fillna("")
df["price"] = pd.to_numeric(df["price"], errors="coerce")
df["average_rating"] = pd.to_numeric(df["average_rating"], errors="coerce")
```

```python
df["title"] = df["title"].str.strip()
df["main_category"] = df["main_category"].str.strip().str.lower()
```

```python
df.to_csv("../data/silver/products_cleaned.csv", index=False)
```
