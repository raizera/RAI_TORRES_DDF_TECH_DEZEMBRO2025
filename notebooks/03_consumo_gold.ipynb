```python
import pandas as pd

df = pd.read_csv("../data/silver/products_cleaned.csv")
df.head()
```

Categorias com mais produtos

```python
df["main_category"].value_counts().head(10)
```

Preço médio por categoria

```python
df.groupby("main_category")["price"].mean().sort_values(ascending=False).head(10)
```

produtos mais bem avaliados

```python
df.sort_values("average_rating", ascending=False)[
    ["title", "main_category", "average_rating"]
].head(10)
```

Comprimento das descrições (IA-friendly)

```python
df["desc_len"] = df["description"].str.len()
df["desc_len"].describe()
```

preparação para IA

Esse campo pode ser usado para:
*embeddings*,
*busca semântica* e
*classificação automática*

```python
df["text_for_ai"] = (
    df["title"].fillna("") + " " + df["description"].fillna("")
)
```

```python
df.to_csv("../data/gold/products_analytics.csv", index=False)
```
