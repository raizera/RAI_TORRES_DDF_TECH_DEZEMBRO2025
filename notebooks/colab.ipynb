```python
import pyarrow.ipc as ipc
import pandas as pd
import glob
import os
import pyarrow as pa

# Caminho dos arquivos Arrow
arrow_files = sorted(glob.glob("data/raw/data-*.arrow"))

print(f"{len(arrow_files)} arquivos encontrados")

tables = []

for file_path in arrow_files:
    with open(file_path, "rb") as f:
        reader = ipc.RecordBatchStreamReader(f)
        table = reader.read_all()
        tables.append(table)

# Concatena todas as tabelas Arrow
full_table = pa.concat_tables(tables)

# Converte para pandas
df = full_table.to_pandas()

print("Shape final:", df.shape)
print("Colunas:", df.columns)

# Amostragem segura
n = min(150_000, len(df))
df_sample = df.sample(n, random_state=42)

# Exporta CSV
df_sample.to_csv("data/raw/products_raw.csv", index=False)

print("CSV gerado:", os.path.exists("products_raw.csv"))
```
